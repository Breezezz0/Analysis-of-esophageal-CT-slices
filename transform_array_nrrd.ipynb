```python
import numpy as np
import nrrd
import nibabel as nib
import os
```

獲得nii file 路徑

```python
nii_file_ct = []
nii_file_path = r'/content/drive/Shareddrives/食道癌/ct_pet_nii(ct pet的nii檔)'#放nii的路徑
for root, dirs, files in os.walk(nii_file_path, topdown=False):
  for filename in files :
    if 'ct' in filename and 'nii' in filename : 
      fullpath = os.path.join(root,filename)  
      if os.path.isfile(fullpath):
        nii_file_ct.append(fullpath)
```

將修改的array存成nrrd

```python
save_nrrd_path = ''#自己設，一個資料夾路徑，結尾不要加/
for nii_file in nii_file_ct :
  img = nib.load(nii_file)
  img_array = img.get_fdata()
  array_transformed = np.where(img_array<-100,-100,img_array)#修改array
  index = nii_file.find('r/')#找原先的路徑要修改的終點
  save_path = nii_file.replace(nii_file[:index+1],save_nrrd_path)#將原先的路徑變成要儲存的路徑
  save_path = save_path.replace('nii.gz','nrrd')
  nrrd.write(save_path, array_transformed)
```
