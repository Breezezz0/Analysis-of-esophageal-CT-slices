```python
params_dir = r'/content/drive/Shareddrives/食道癌/pyradiomics/exampleSettings/'
params_name = os.listdir(params_dir)
params_fullpath = []
for i in params_name :
  fullpath = params_dir + i
  params_fullpath.append(fullpath)
for params in params_fullpath :
  image_path_dir=("/content/drive/Shareddrives/食道癌/pet_nrrd_0-64")
  label_path_dir=("/content/drive/Shareddrives/食道癌/local_re_計算rad程式碼和結果/roi_nrrd_new")
  roi_file_list = os.listdir(label_path_dir)
  pet_file_list = os.listdir(image_path_dir)
  id=[]
  for f in roi_file_list:
      id.append(f[:-5])
  for i in id:
      for ii in roi_file_list:
          if i in ii:
              label_path = label_path_dir + "/" + ii
      for iii in pet_file_list:
          if i in iii:
            pet_file_id = image_path_dir + "/" + iii 
            pet_file_id_list=os.listdir(pet_file_id)
            for k in pet_file_id_list:
              if "pet" in k:
                image_path = pet_file_id + "/" + k
      try:  
        extractor = featureextractor.RadiomicsFeatureExtractor(params)
        result = extractor.execute(image_path,label_path) # type is dict
        key_name=[]
        value_str=[]
        for key, value in result.items():
          key_name.append(key)
          value_str.append(str(value))
        df = pd.DataFrame({'value':value_str},
                    index=key_name,
                    columns=['value'])
        save_dir = r'/content/drive/Shareddrives/食道癌/rad_pet_0-64/'
        save_subdir = save_dir + params.split('/')[-1]
        if not os.path.isdir(save_subdir) :
          os.makedirs(save_subdir)
        df.to_excel(save_subdir +'/'+str(i)+'.xlsx')
      except:
        print(image_path)
        print(label_path)
        continue
```
