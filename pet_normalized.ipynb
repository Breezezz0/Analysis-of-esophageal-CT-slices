```python
import numpy as np
import os
import pandas as pd
import nrrd
import pydicom
```

```python
nrrd_path_dir = '/content/drive/Shareddrives/食道癌/ct_pet_nrrd(ct pet的nrrd檔)'
nrrd_file = []
for root , dir , files in os.walk (nrrd_path_dir , topdown=False) :
  for filename in files :
    if 'pet' in filename and 'nrrd' in filename :
      fullpath = os.path.join(root,filename)
      if os.path.isfile(fullpath):
        nrrd_file.append(fullpath)
```

```python
print(nrrd_file)
```

```python
new_nrrd_path = '/content/drive/Shareddrives/食道癌/pet_正規化的檔案'
for file in nrrd_file :
  array , header = nrrd.read(file)
  min = 0
  max = np.max(array)
  array = array.astype("float32")
  filter_b = tuple([array > max])
  array[array<0] = 0
  array = ((array - min) / (max - min)) * 64
  array[filter_b] = 64
  index = file.find('3/')#找原先的路徑要修改的終點
  save_path = file.replace(file[:index+1],new_nrrd_path)#將原先的路徑變成要儲存的路徑
  last_dir_index = save_path.rfind('/')
  if not os.path.isdir(save_path[:last_dir_index+1]):
    os.makedirs(save_path[:last_dir_index+1])
  nrrd.write(save_path , array)
```
